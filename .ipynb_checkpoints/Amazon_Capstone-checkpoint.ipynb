```python
df["Ship State"] = df["Ship State"].str.upper()
df["Ship State"].replace({"PUDUCHERRY":"PONDICHERRY","RAJSHTHAN":"RAJASTHAN","RAJSTHAN":"RAJASTHAN","RJ":"RAJASTHAN",
                         "PB":"PUNJAB","PUNJAB/MOHALI/ZIRAKPUR":"PUNJAB","ORISSA":"ODISHA","DELHI":"NEW DELHI",
                         "NL":"UNKNOWN","APO":"UNKNOWN","AR":"UNKNOWN"}, inplace = True)
```

```python
# creating target column as reject and not-reject

def Status(x):
    if x in ('Cancelled','Shipped - Returning to Seller','Shipped - Rejected by Buyer','Shipped - Returned to Seller'):
        return 'Rejected'
    elif x in ('Shipped', 'Shipped - Delivered to Buyer', 'Shipped - Picked Up', 'Shipped - Out for Delivery','Pending - Waiting for Pick Up', 'Pending'):
        return 'Not Rejected'
    else:
        return '0'
```

```python
df["Status"].value_counts()
```

```python
df["Status"] = df["Status"].apply(Status)
```

```python
# Shipping, Shipped - Lost in Transit, Shipped - Damaged --- not sure if it will be cancelled or not, 
# so grouped it as 0 and dropping these records.

df["Status"].value_counts()
```

```python
df[df["Status"]=='0'].index
```

```python
df.drop(index=[  4017,   4023,  52733,  68542,  68543, 102676, 107949, 107951,
            107953, 107954, 107955, 107957, 107959, 107960],axis=0,inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
# few records of Amount column is 0 but the status is not rejected, hence only for not rejected records fill the 0 
# with meandian of that category type.
```

```python
gp=df.groupby("Category")["Amount"].median()
```

```python
a=gp.index
```

```python
for index, row in df.iterrows():
    if row["Status"] == "Not Rejected":
        if row["Amount"] == 0.0:
            for key, value in d.items():
                if row['Category'] in key:
                    df.at[index, "Amount"] = value
```

```python
df[df["Amount"]==0.0].shape
```

```python
# only rejected columns are left as 0 amount

df[(df["Status"]=='Rejected') & (df["Amount"]==0.0)].shape
```

```python
# filling null values of Courier Status column

df['Courier Status'].unique()
```

```python
for index, row in df.iterrows():
    if pd.isnull(row["Courier Status"]):
        if row["Status"] == 'Rejected':
            df.at[index, "Courier Status"] = 'Cancelled'
        else:
            df.at[index, "Courier Status"] = 'Shipped'
```

```python
# grouping state columns, for further analysis

def State(x):
    if x in ('UTTAR PRADESH','NEW DELHI','HARYANA','JAMMU & KASHMIR','CHANDIGARH','UTTARAKHAND','LADAKH','HIMACHAL PRADESH'):
        return 'North India'
    elif x in ('MAHARASHTRA','KARNATAKA','TAMIL NADU','TELANGANA','KERALA','ANDHRA PRADESH','PONDICHERRY'):
        return 'South India'
    elif x in ('WEST BENGAL','MIZORAM','ASSAM','JHARKHAND','SIKKIM','TRIPURA','CHHATTISGARH','ODISHA','MADHYA PRADESH',
               'MEGHALAYA','NAGALAND','BIHAR','MANIPUR','ARUNACHAL PRADESH'):
        return 'East India'
    elif x in ('RAJASTHAN','GUJARAT','PUNJABHIMACHAL PRADESH','GOA','DADRA AND NAGAR'):
        return 'West India'
    else:
        return 'Others'
```

```python
df["State_New"]=df["Ship State"].apply(State)
```

```python
df_cat.drop(['SKU','promotion-ids'], axis=1, inplace = True)
```

```python
df_cat1 = pd.get_dummies(df_cat[['Fulfilment','Sales Channel','Ship Service Level','Category','Courier Status','State_New']], drop_first=True)
```

```python
# Dummy =  Fulfilment , Sales Amazon, Ship Service Level, Category, Courier Status, 
# Ordinal = Size
```

```python
df1['Size'].unique()
```

```python
oe = OrdinalEncoder(categories=[['XS','S','M','L','XL','XXL','3XL','4XL','5XL','6XL', 'Free']])
size = oe.fit_transform(df_cat[['Size']])
df_cat2 = pd.DataFrame(size)
```

```python
df_cat2.rename(columns = {0:'Size'}, inplace=True)
```

```python
df2 = pd.concat([df_num, df_cat1], axis=1)
```

```python
df_cat2 = df_cat2.set_index(df['Order ID'])
```

```python
final = pd.concat([df2, df_cat2], axis=1)
```

```python
X = final.drop('Status', axis=1)
y = final['Status']
```

```python
X.shape
```

```python
y.shape
```

```python
Xc = sm.add_constant(X)
```

```python
model = sm.Logit(y,Xc).fit()
model.summary()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print('Test Accuracy:', accuracy)
```

```python
train_accuracy = knn.score(X_train, y_train)
print('Train Accuracy:', train_accuracy)
```
